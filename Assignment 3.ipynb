```python

# match statement:
#The match statement is used to compare a value with multiple patterns and execute the matching block.
#It works like switch-case in other languages and is cleaner than long if/elif chains.
```

```python
day = 3

match day:
    case 1:
        print("Monday")
    case 2:
        print("Tuesday")
    case 3:
        print("Wednesday")
    case 4:
        print("Thursday")
    case 5:
        print("Friday")
    case _:
        print("Invalid day")
```

```python
a = 10
b = 5
op = "+"

match op:
    case "+":
        print(a + b)
    case "-":
        print(a - b)
    case "*":
        print(a * b)
    case "/":
        print(a / b)
    case _:
        print("Invalid operator")
```

```python
grade = "C"

match grade:
    case "A":
        print("Excellent")
    case "B":
        print("Good")
    case "C":
        print("Average")
    case "D":
        print("Poor")
    case _:
        print("Fail")
```

```python
"""The if/else statement is used to check conditions that are either true or false, such as comparisons and logical decisions.
It is useful when different conditions ""need to be tested using operators. On the other hand, the match statement is used to match a single value with multiple fixed cases. It makes the code easier to read when there are many options. The if/else statement works in all versions of Python, while the match statement works only in Python 3.10 and above. In simple words, if/else is used for condition checking, whereas match is used for choosing from fixed values"""
```

```python
# Example 1 of nested if else
num = 10

if num >= 0:
    if num == 0:
        print("Zero")
    else:
        print("Positive")
else:
    print("Negative")
```

```python

# Example
marks = 75

if marks >= 40:
    if marks >= 80:
        print("Grade A")
    else:
        print("Pass")
else:
    print("Fail")
```
